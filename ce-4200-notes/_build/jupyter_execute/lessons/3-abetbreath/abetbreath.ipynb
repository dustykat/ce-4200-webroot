# ABET Breadth of Knowledge Considerations for Civil Engineering

## **Introduction (5 minutes)**

### **Opening Hook**
- *"What makes a Civil Engineer's education complete?"*

### **Objective**
- Introduce the importance of breadth in knowledge as required by ABET.
- Emphasize how it prepares students for multidisciplinary challenges in civil engineering.

### **Key Points**
- Overview of ABET accreditation and its role in civil engineering programs.
- The significance of breadth in knowledge for career readiness and ethical responsibility.

---

## **Part 1: Core Technical Knowledge (10 minutes)**

### **Definition**
- Explain how ABET defines the technical knowledge required for civil engineers.

### **Discussion**
- **Fundamental Areas**:
  - Structural, geotechnical, water resources, transportation, environmental, and construction engineering.
- **Integrative Design**:
  - Importance of combining these areas in project-based learning.
- **Case Study**:
  - Example: A dam project requiring hydrology, structural integrity, and environmental impact assessment.

---

## **Part 2: Supporting Knowledge Areas (10 minutes)**

### **Mathematics and Natural Sciences**
- Emphasize calculus, differential equations, physics, and chemistry as foundations.
- Highlight application in real-world problem-solving.

### **Broader Technical Skills**
- Programming and computational tools (e.g., MATLAB, Python, GIS).
- Emerging technologies like AI and sustainability tools.

### **Humanities and Social Sciences**
- Importance of understanding societal impact, policy, and communication.
- Example: Integrating public safety and welfare in transportation planning.

---

## **Part 3: Professional Skills and Ethical Responsibility (10 minutes)**

### **Interpersonal and Communication Skills**
- Importance of teamwork and leadership.
- **Exercise Suggestion**:
  - Role-playing: Conflict resolution in a multidisciplinary team.

### **Ethics in Engineering**
- Overview of NSPE Code of Ethics.
- **Ethical Dilemma Scenario**:
  - Balancing budget constraints with safety requirements.

### **Global and Societal Context**
- Projects like infrastructure design have far-reaching consequences.
- **Example**:
  - Sustainable urban planning.

---

## **Conclusion (5 minutes)**

### **Recap**
- ABET's vision for a well-rounded engineer.
- Breadth of knowledge as a cornerstone of civil engineering education.

### **Call to Action**
- Encourage students to seek interdisciplinary experiences during their studies.
- Highlight lifelong learning as a professional imperative.
