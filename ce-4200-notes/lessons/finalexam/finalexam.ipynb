## Introduction to the Final Exam

The final exam is comprised of questions derived from the "Framing the Profession" section, the "Beyond the Technical" section, and an FE Exam simulation (Mock FE - reduced length and scope).  

The FE exam simulation is based on circa 2017 and earlier materials.  You are allowed the [NCEES supplied reference](https://help.ncees.org/article/87-ncees-exam-reference-handbooks) and a calculator (the embedded calculator in Canvas is turned on, but be careful to clear work between problems).

:::{note}
Using FE specifications as a guideline, the exam will take a bit over 3 hours to complete (2 minutes/question for FE questions, 1 minutes/question for profesional practices contents).  As such the exam is provided online via Canvas, with several days to work through the exam.  A preview of the exam is linked below.
:::
